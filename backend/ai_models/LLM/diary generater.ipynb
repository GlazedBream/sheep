```python
# â”€â”€â”€ ê¸°ë³¸ ì„¤ì • â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
import os
import io
import json
import base64
from datetime import datetime
from dotenv import load_dotenv

# â”€â”€â”€ ì´ë¯¸ì§€ ë° EXIF ì²˜ë¦¬ â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
from PIL import Image
import exifread

# â”€â”€â”€ OpenAI ë° LangChain â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
import openai
from openai import OpenAI

# LangChain ê´€ë ¨ ëª¨ë“ˆ
from langchain_openai import ChatOpenAI
from langchain.prompts.chat import ChatPromptTemplate
from langchain.prompts import PromptTemplate
from langchain.schema import SystemMessage, HumanMessage
```

```python


# í™˜ê²½ ë³€ìˆ˜ ë¡œë“œ (.envì— OPENAI_API_KEY í¬í•¨ë˜ì–´ ìˆì–´ì•¼ í•¨)
load_dotenv()
client = OpenAI()
chat_model = ChatOpenAI(temperature=0.9)

# ì´ë¯¸ì§€ ì••ì¶• í•¨ìˆ˜ (base64 ë³€í™˜ í¬í•¨)
def compress_image(image_path, size=(512, 512)):
    img = Image.open(image_path)
    img = img.resize(size)
    buffer = io.BytesIO()
    img.save(buffer, format="JPEG")
    return base64.b64encode(buffer.getvalue()).decode("utf-8")

# ì´ë¯¸ì§€ ì´¬ì˜ ì‹œê°„ ê¸°ì¤€ ì •ë ¬ (EXIF ì‚¬ìš©)
def extract_datetime_from_image(image_path):
    with open(image_path, 'rb') as f:
        tags = exifread.process_file(f, stop_tag="EXIF DateTimeOriginal")
    datetime_str = tags.get("EXIF DateTimeOriginal")
    if datetime_str:
        return datetime.strptime(str(datetime_str), "%Y:%m:%d %H:%M:%S")
    return None

def sort_all_images(image_folder):
    images = []
    for filename in os.listdir(image_folder):
        if filename.lower().endswith(('.jpg', '.jpeg')):
            path = os.path.join(image_folder, filename)
            dt = extract_datetime_from_image(path)
            if dt:
                images.append((filename, dt))
    return sorted(images, key=lambda x: x[1])

# ìº¡ì…˜ ìºì‹œ ë¡œë”© ë° ì €ì¥
def load_cached_captions(cache_path):
    if os.path.exists(cache_path):
        with open(cache_path, 'r', encoding='utf-8') as f:
            return json.load(f)
    return {}

def save_cached_captions(cache_path, data):
    with open(cache_path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)

# ì´ë¯¸ì§€ â†’ ìº¡ì…˜ ìƒì„± (OpenAI GPT-4 Vision ì‚¬ìš©)
def image_to_caption(image_path, prompt="ì´ ì´ë¯¸ì§€ë¥¼ ì‚¬ì‹¤ì ìœ¼ë¡œ ë¬˜ì‚¬í•´ì¤˜."):
    b64_img = compress_image(image_path)
    response = client.chat.completions.create(
        model="gpt-4-turbo-2024-04-09",
        messages=[
            {"role": "system", "content": "ë„ˆëŠ” ì‚¬ì‹¤ì ì¸ ì´ë¯¸ì§€ ì„¤ëª…ì„ ì˜ í•˜ëŠ” ì‘ê°€ì•¼."},
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": prompt},
                    {"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{b64_img}"}},
                ],
            },
        ],
        max_tokens=300
    )
    return response.choices[0].message.content

# LangChain ë‹¤ì´ì–´ë¦¬ í”„ë¡¬í”„íŠ¸
diary_prompt = ChatPromptTemplate.from_messages([
    ("system", "ë„ˆëŠ” ê°ì„±ì ì¸ ì¼ê¸° ì‘ê°€ì•¼."),
    ("human", 
    """ì•„ë˜ ì‚¬ì§„ ì„¤ëª…ë“¤ì„ ë°”íƒ•ìœ¼ë¡œ í•˜ë£¨ë¥¼ ê°ì„±ì ìœ¼ë¡œ ì •ë¦¬í•´ì¤˜.

- ì‚¬ì§„ ì„¤ëª…ë“¤: {image_caption}
- ê°ì •: {emotion}
- í‚¤ì›Œë“œ: {keywords}
- ì¥ì†Œ: {place}

ì¡°ê±´: ë„ˆë¬´ í˜•ì‹ì ì´ì§€ ì•Šê³ , ìì—°ìŠ¤ëŸ½ê³  ë”°ëœ»í•œ 300ìë¡œ ì¨ì¤˜.""")
])

# ì „ì²´ íŒŒì´í”„ë¼ì¸ ì‹¤í–‰
folder_path = r"C:\Users\Admin\Desktop\í”„ë¡œì íŠ¸\ì¼ê¸°ìƒì„±\test"  # ì´ë¯¸ì§€ í´ë” ê²½ë¡œ
cache_path = os.path.join(folder_path, "caption_cache.json")

sorted_images = sort_all_images(folder_path)
caption_cache = load_cached_captions(cache_path)

all_captions = []
for filename, dt in sorted_images:
    if filename in caption_cache:
        caption = caption_cache[filename]
    else:
        image_path = os.path.join(folder_path, filename)
        caption = image_to_caption(image_path)
        caption_cache[filename] = caption
    all_captions.append(caption)

# ìºì‹œ ì €ì¥
save_cached_captions(cache_path, caption_cache)

# ì „ì²´ ìº¡ì…˜ì„ í•˜ë‚˜ì˜ ë¬¸ì¥ìœ¼ë¡œ í•©ì¹¨
combined_caption = " ".join(all_captions)

# LangChain í”„ë¡¬í”„íŠ¸ êµ¬ì„±
messages = diary_prompt.format_messages(
    image_caption=combined_caption,
    emotion="ì•„ì£¼ í–‰ë³µí–ˆë˜ í•˜ë£¨ì˜€ì–´",  # ì¶”í›„ ìë™ ê°ì • ì¶”ë¡  ê°€ëŠ¥
    keywords="í–‡ì‚´, ì¡°ìš©í•œ ê±°ë¦¬, ì—¬í–‰",
    place=" "
)

# ì¼ê¸° ìƒì„±
diary = chat_model.invoke(messages).content

# ê²°ê³¼ ì¶œë ¥
print("ì „ì²´ ì´ë¯¸ì§€ ê¸°ë°˜ í•˜ë£¨ ìš”ì•½ ì¼ê¸°")
print(diary)
```
